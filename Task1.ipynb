```python
import pandas as pd
ds=pd.read_csv('world_population1.csv').head(40)
data={
    'rank':ds['Rank'],
    'capital':ds['Capital'],
    'country':ds['Country'],
    'continent':ds['Continent'],
    '2020 population':ds['2020 Population']
}
df=pd.DataFrame(data)
#identify duplicate rows 
duplicates=df.duplicated()
print("Duplicate rows in bool")
print(duplicates)
#Duplicate rows
duplicate=df[df.duplicated()]
print("\nDuplicate rows:")
duplicate
```

```python
df_no_duplicates_groupby=df.groupby('country').first().reset_index()
print("\nData using groupby and first():")
df_no_duplicates_groupby
```

```python
pivot_tables=df.pivot_table(index=['capital'],aggfunc='first')
print("\npivot table after removing duplicates using pivot_table and first():")
pivot_tables
```

```python
import pandas as pd
import matplotlib.pyplot as plt
ds=pd.read_csv('world_population1.csv').head(40)
data={
    'rank':ds['Rank'],
    'capital':ds['Capital'],
    'country':ds['Country'],
    'continent':ds['Continent'],
    '2020 population':ds['2020 Population']
}
df=pd.DataFrame(data)
df_unique=df.drop_duplicates()
plt.figure(figsize=(12,6))

#plot original data
plt.subplot(1,2,1)
plt.plot(df.index,df['2020 population'],marker='o',linestyle='-',color='b')
plt.title('Original value')
plt.xlabel('Index')
plt.ylabel('Age')
plt.grid(True)

#plot duplicate data
plt.subplot(1,2,2)
plt.plot(df_unique.index,df_unique['2020 population'],marker='o',linestyle='-',color='r')
plt.title('data after removing duplicates')
plt.xlabel('Index')
plt.ylabel('Age')
plt.grid(True)

plt.tight_layout()
plt.show()
    
```

```python
from scipy.stats import norm
import numpy as np
import statistics
ds=pd.read_csv('world_population1.csv').head(40)
data={
    'rank':ds['Rank'],
    'capital':ds['Capital'],
    'country':ds['Country'],
    'continent':ds['Continent'],
    '2020 population':ds['2020 Population']
}
n=ds['Rank']
x=statistics.mean(n)
y=statistics.stdev(n)
score=60
z_score=(score-x)/y
prob=norm.cdf(z_score)
percent=prob*100
print("percentage is",round(percent,2),"%")
print(len(n))
```

```python
import pandas as pd
dataset=pd.read_csv('world_population1.csv').head(40)
num=dataset['Rank']
def find_mean(list_of_num):
    sum_n=sum(list_of_num)
    len_n=len(list_of_num)
    mean=sum_n/len_n
    return mean
result=find_mean(num)
print(result)
```
