```python
import pandas as pd
import numpy as np
dataset=pd.read_csv('Crimes_in_india_2001-2013.csv').head(150)
data={
    'DISTRICT':dataset['CHEATING'],
    'STATE/UT':dataset['HURT/GREVIOUS HURT']
     }
df=pd.DataFrame(data)
z_score=np.abs((df-df.mean())/df.std())
thershold=2
outliers=z_score>thershold
df_no_outliers=df[(z_score<thershold).all(axis=1)]
print("original dataframe")
print(df)
print("\nz_score:")
print(z_score)
print("\nOutliers detected:")
print(outliers)
print("\nDataframe after removing:")
print(df_no_outliers)
```

```python
data1=df_no_outliers
data1
data1.plot(kind='bar',color='green')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
dataset=pd.read_csv('Crimes_in_india_2001-2013.csv').tail(60)
data={
    'DISTRICT':dataset['CHEATING'],
    'STATE/UT':dataset['ROBBERY']
     }
df=pd.DataFrame(data)
df_unique=df.drop_duplicates()
plt.figure(figsize=(12,6))

#plot original data
plt.subplot(1,2,1)
plt.plot(df.index,df['STATE/UT'],marker='o',linestyle='-',alpha=0.8,color='orange')
plt.title('Original value')
plt.xlabel('Index')
plt.ylabel('Age')
plt.grid(True)

#plot duplicate data
plt.subplot(1,2,2)
plt.plot(df_unique.index,df_unique['STATE/UT'],marker='o',alpha=0.8,linestyle='-',color='purple')
plt.title('data after removing duplicates')
plt.xlabel('Index')
plt.ylabel('Age')
plt.grid(True)

plt.tight_layout()
plt.show()
    
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('Crimes_in_india_2001-2013.csv')
data.groupby('YEAR')['TOTAL IPC CRIMES'].mean().plot(kind='line',color='blue')
plt.title('Average Crime Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Average Crime Rate')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
dataset=pd.read_csv('Crimes_in_india_2001-2013.csv').head(5000)
data={
    'other crimes':dataset['OTHER IPC CRIMES'],
    'state':dataset['YEAR'],
    'total crime':dataset['TOTAL IPC CRIMES'],
    'murder':dataset['MURDER']   
}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
print(correlation_matrix)
import seaborn as sns
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='inferno', vmin=-1, vmax=1, fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()

```
