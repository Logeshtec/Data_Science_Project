```python
import pandas as pd
dataset=pd.read_csv('ipl_match_info_data.csv')
headvalue=dataset.head(70)
num=headvalue['results']
def find_mean(list_of_num):
    sum_n=sum(list_of_num)
    len_n=len(list_of_num)
    mean=sum_n/len_n
    return mean
result=find_mean(num)
print(result)
```

```python
import pandas as pd
dataset=pd.read_csv('ipl_match_info_data.csv').head(70)
n=dataset['results']
def calculate_mode(list_of_n):
    counter={}
    for i in list_of_n:
        if i in counter:
            counter[i]+=1
        else:
            counter[i]=1
    k_max=sorted(counter,key=counter.get,reverse=True)[18]
    frequency=counter[k_max]
    return k_max,frequency
print(calculate_mode(n))
        
```

```python
import pandas as pd
import matplotlib.pyplot as plt
dataset=pd.read_csv('ipl_match_info_data.csv').head(46)
data={
    'type':dataset['result_type'],
    'runs':dataset['winner_runs'],
    'umpire':dataset['umpire1'],
    'venue':dataset['venue']
}
df=pd.DataFrame(data)
df_unique=df.drop_duplicates()
plt.figure(figsize=(12,6))

#plot original data
plt.subplot(1,2,1)
plt.plot(df.index,df['runs'],marker='o',linestyle='-',color='b')
plt.title('Original value')
plt.xlabel('Index')
plt.ylabel('Age')
plt.grid(True)

#plot duplicate data
plt.subplot(1,2,2)
plt.plot(df_unique.index,df_unique['runs'],marker='o',linestyle='-',color='r')
plt.title('data after removing duplicates')
plt.xlabel('Index')
plt.ylabel('Age')
plt.grid(True)

plt.tight_layout()
plt.show()
    
```

```python
import pandas as pd
import numpy as np
dataset=pd.read_csv('ipl_match_info_data.csv').head(70)
data={
    'type':dataset['results'],
    'runs':dataset['winner_runs']
}
df=pd.DataFrame(data)
z_score=np.abs((df-df.mean())/df.std())
thershold=3
outliers=z_score>thershold
df_no_outliers=df[(z_score<thershold).all(axis=1)]
print("original dataframe")
print(df)
print("\nz_score:")
print(z_score)
print("\nOutliers detected:")
print(outliers)
print("\nDataframe after removing:")
print(df_no_outliers)
```
